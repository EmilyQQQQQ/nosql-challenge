5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude

# Define the query to find documents with latitude and longitude as strings
query = {
    "$or": [
        {"latitude": {"$type": "string"}},
        {"longitude": {"$type": "string"}}
    ]
}

# Define the update to convert latitude and longitude to Decimal
update = {
    "$set": {
        "latitude": {"$convert": {"input": "$latitude", "to": "decimal"}},
        "longitude": {"$convert": {"input": "$longitude", "to": "decimal"}}
    }
}

# Use update_many to update the data type of latitude and longitude
LatLonUpdate = collection.update_many(query, update)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]

# Set non-rating values to null (None)
query = {"RatingValue": {"$in": non_ratings}}
update = {"$set": {"RatingValue": None}}
result = collection.update_many(query, update)
```

```python
# Change the data type from String to Integer for RatingValue
query = {"RatingValue": {"$type": "string"}}
update = {"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}
RatingValueUpdate = collection.update_many(query, update)
```
